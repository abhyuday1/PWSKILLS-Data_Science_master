## Q1: What are the Probability Mass Function (PMF) and Probability Density Function (PDF)? Explain with an example.

-- Probability Mass Function :
--
A PMF is used to describe the probability distribution of a discrete random variable. It gives the probability of the random variable assuming each of its possible values. In other words, it maps each possible value of the random variable to its corresponding probability.

Mathematically for discrete Random Variable X PMF is P(X=x)
x is specific value and satisfies the following properties:-
1. P(X=x) >=0 for all possible values of x.
2. summation of P(X=x)=1, where sum is taken over all possible vaues of X.

Example:-

Consider a random Variable  X representing the outcome of rolling fair dice.The PMF for X are:-

P(X=1)=1/6

P(X=2)=1/6

P(X=3)=1/6

P(X=4)=1/6

P(X=5)=1/6

P(X=6)=1/6

-- Probability Density Function
--

A PDF is used to describe the probability distribution of a continuous random variable. Unlike the PMF, which deals with discrete values, the PDF deals with continuous values.

Mathematically, for a continuous random variable X the PDF is denoted as f(x) where
x is a specific value, and it satisfies the following properties:

1. f(x) >=0 for all possible values of x.
2. integral(f(x)dx=1, where integral is taken over all possible vaues of x.

 Example
 
 Consider the random variable  X representing the height of individuals in a population, which follows a normal distribution.
 
 In this case, the PDF provides the likelihood of an individual's height falling within a specific range. For example, it can tell us the probability of selecting an individual with a height between 160 cm and 170 cm from the population.


## Q2: What is Cumulative Density Function (CDF)? Explain with an example. Why CDF is used?

-- Cumulative Density Function is used to describe the cumulative probability of a random variable taking on a value less than or equal to a specific value

Mathematically  for a random variable X the CDF is Denoted as F(x) where x is specific value and defined as:

F(x)=P(X<=x)


the CDf satisfies the following properties:-
1. F(X) is non-decreasing: Either  stays the same or increases.
2. lim (-infinty)F(X)=0  and lim(infinty)F(X)=1

Example:-
Consider a six-sided rolling dice. the CDF forthis random variable is denoted as F(x), can be expressed as :
F(X)=0  for x < 1

F(X) =1/6 for 1<=x <2
F(X) =1/6 for 2<=x <3
F(X) =1/6 for 3<=x <4
F(X) =1/6 for 4<=x <5
F(X) =1/6 for 5<=x <6
F(X) =1 for x>6
-- the CDF tells us the cumulative probability of rolling a value less than or equal to a specific number on the die.

--Why CDF is Used
--

1. It allows to find the probability that a random variable falls within a particular range, which is useful for understanding the likelihood of various events occurring.

2. It helpsto  find percentiles and quantiles of a distribution, which are crucial in statistics for making inferences and setting thresholds.

3.  It is often used in hypothesis testing and statistical analysis.
