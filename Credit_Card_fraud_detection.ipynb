```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=42)
```

```python
type(x_train)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
test_ss=StandardScaler()
train_ss=StandardScaler()
```

```python
ss_x_train=train_ss.fit_transform(x_train)
#ss_x_test=test_ss.fit_transform(x_test)
```

```python
ss_x_test=test_ss.fit_transform(x_test)
```

```python
scaled_xtrain=pd.DataFrame(ss_x_train,columns=x_train.columns,index=x_train.index)
```

```python
scaled_xtest=pd.DataFrame(ss_x_test,columns=x_test.columns,index=x_test.index)
```

```python
x_test.columns
```

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import RandomForestClassifier
```

```python
model_nb=GaussianNB()
```

```python
model_nb.fit(scaled_xtrain,y_train)
```

```python
model_rf=RandomForestClassifier()
```

```python
model_rf.fit(scaled_xtrain,y_train)
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
```

```python
y_pred_nb=model_nb.predict(scaled_xtest)
```

```python
y_pred_rf=model_rf.predict(scaled_xtest)
```

```python
print(f'Accuracy score for Naive Bayes : {accuracy_score(y_test,y_pred_nb)}')
```

```python
print(f'Accuracy score for Naive Bayes : {accuracy_score(y_test,y_pred_rf)}')
```

```python
print(confusion_matrix(y_test,y_pred_nb))
```

```python
print(classification_report(y_test,y_pred_nb))
```

```python
print(confusion_matrix(y_test,y_pred_rf))
```

```python
print(classification_report(y_test,y_pred_rf))
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid={'var_smoothing':[0.1,0.01,0.5,0.005,0.0001]}
```

```python
grid=GridSearchCV(estimator=model_nb,param_grid=param_grid,cv=5, verbose=3)
```

```python
grid.fit(scaled_xtrain,y_train)
```

```python
grid.best_params_
```

```python
grid.best_estimator
```

```python
model_nb_grid=GaussianNB(var_smoothing=0.5)
```

```python
model_nb_grid.fit(scaled_xtrain,y_train)
```

```python
y_pred_nbgrid=model_nb_grid.predict(scaled_xtest)
```

```python
print(f'Accuracy score for Naive Bayes: {accuracy_score(y_test,y_pred_nbgrid)}')
```

```python
print(confusion_matrix(y_test,y_pred_nbgrid))
```

```python
print(classification_report(y_test,y_pred_nbgrid))
```

```python
param_grid_rf={'n_estimators': [50,100,150],
              'max_depth':range(3,11,1),
               "random_state" :[0,50,100],
               "criterion": ['gini','entropy']
              }
```

```python
grid_rf=GridSearchCV(model_rf,param_grid=param_grid_rf,cv=5,verbose=3)
```

```python
grid_rf.fit(scaled_xtrain,y_train)
```

```python

#print(grid_rf.best_estimator)
print(grid_rf.best_params_)

model_rf_grid=RandomForestClassifier(**grid_rf.best_params_)
model_rf_grid.fit(scaled_xtrain,y_train)
y_pred_rfgrid=model_rf_grid.predict(scaled_xtest)
print(f'Accuracy score for Random forest: {accuracy_score(y_test,y_pred_rfgrid)}')
print(confusion_matrix(y_test,y_pred_rfgrid))
print(classification_report(y_test,y_pred_rfgrid))
```
