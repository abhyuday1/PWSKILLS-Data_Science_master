## Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data plot a scatter plot.

import numpy as np

np.random.seed(3)

x = 3 + np.random.normal(0, 2, 50)

y = 3 + np.random.normal(0, 2, len(x))

Note: Also add title, xlabel, and ylabel to the plot.

```python
'''A scatter plot is a type of data visualization that displays individual data points as markers on a
two-dimensional plane, where the position of each point is determined by its values along two variables,
typically representing two different attributes or features.'''

import numpy as np
import matplotlib.pyplot as plt
np.random.seed(3)

x = 3 + np.random.normal(0, 2, 50)

y = 3 + np.random.normal(0, 2, len(x))
```

```python
plt.scatter(x,y)
plt.title("Scatter Plot for X VS Y")
plt.xlabel("X-axis or x- value")
plt.ylabel("Y-axis or Y-Value")
```

## Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.
#### Use the following data:

import numpy as np

For line 1: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([0, 100, 200, 300, 400, 500])

For line 2: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([50, 20, 40, 20, 60, 70])

For line 3: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60])

For line 4: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150])


```python
'''The subplot() function in Matplotlib is used to create multiple plots within the same figure, arranging them in a grid.
This allows you to display multiple plots or visualizations in a single figure'''

import numpy as np
import matplotlib.pyplot as plt


x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])

x2 = np.array([0, 1, 2, 3, 4, 5]) 
y2 = np.array([50, 20, 40, 20, 60, 70])

x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])

x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])


plt.subplot(2,2,1)
plt.plot(x1,y1)
plt.title("Plot for Line-1")

plt.subplot(2,2,2)
plt.plot(x2,y2)
plt.title("Plot for Line-2")

plt.subplot(2,2,3)
plt.plot(x3,y3)
plt.title("Plot for Line-3")

plt.subplot(2,2,4)
plt.plot(x4,y4)
plt.title("Plot for Line-4")

plt.tight_layout() ## adjusting the subplots
```

## Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.

import numpy as np

company = np.array(["Apple", "Microsoft", "Google", "AMD"])

profit = np.array([3000, 8000, 1000, 10000])

```python
'''A bar plot is a data visualization technique used to represent categorical data with rectangular bars.
Each bar's length or height is proportional to the value it represents,allowing for easy comparison of different 
categories or groups.'''

import numpy as np
import matplotlib.pyplot as plt

company = np.array(["Apple", "Microsoft", "Google", "AMD"])

profit = np.array([3000, 8000, 1000, 10000])

plt.bar(company,profit)
plt.title("Companies Profit")
plt.xlabel("Companies")
plt.ylabel("Profit")
```

```python
plt.barh(company,profit)
plt.title("Companies Profit")
plt.xlabel("Companies")
plt.ylabel("Profit")
```

## Q5: What is a box plot? Why is it used? Using the following data plot a box plot.

box1 = np.random.normal(100, 10, 200)

box2 = np.random.normal(90, 20, 200)

```python
'''A box plot is a data visualization technique used to display the distribution, central tendency, and spread of a dataset.
It is particularly useful for visualizing the summary statistics and identifying potential outliers in the data'''

box1 = np.random.normal(100, 10, 200)

box2 = np.random.normal(90, 20, 200)

plt.boxplot([box1,box2],labels=['Box1','Box2'])
plt.title('Box plot of Box1 VS Box2')

plt.xlabel("Box")
plt.ylabel("values")
```
