-Custom exceptions are created by defining new classess that inherit from the builtin Exception class.

-Exception handling mechanism including 'try','except' and 'finally' blocks are designed to work with exceptions that inherit from 'Exception' class.

## Q2. Write a python program to print Python Exception Hierarchy.


```python
def print_exception_hierarchy(base_class, indent=0):
    print(' ' * indent + base_class.__name__)
    for subclass in base_class.__subclasses__():
        print_exception_hierarchy(subclass, indent + 3)

if __name__ == "__main__":
    print("Python Exception Hierarchy:")
    print_exception_hierarchy(BaseException)
```

## Q3. What errors are defined in the ArithmeticError class? Explain any two with an example.


--ArithmeticError class is a base class for exceptions related to arithmetic operations. it serves as the parent class for several arithmetic related exceptions.

examples:-

1. ZeroDivisionErorr:-
        ZeroDivisionError is raised when any number is divided by zero. this error occurs because  division by zero is mathematically undefined.

```python
num1 = 8
num2 = 0
try:
    result = num1 / num2  # Attempting to divide by zero
except ZeroDivisionError as e:
    print(f"Error: {e}")
```

2. FloatingPointError:-
    This exception is raise when a floating number is divided by zero.

```python
num1 = 8.98
num2 = 0
try:
    result = num1 / num2  # Attempting to divide by zero
except ZeroDivisionError as e:
    print(f"Error: {e}")
```

## Q4. Why LookupError class is used? Explain with an example KeyError and IndexError.

--lookupError class is used to represent the exceptions occurs while accessing elements in a sequence. or while performing lookup operations.

1. KeyError:- it is raised while accessing a dictionary element using key that does not exist.

```python
my_dict = {'cow': 5, 'goat': 3, 'horse': 8}

try:
    value = my_dict['apple'] 
except KeyError as e:
    print(f"Error: {e}")
```

2. IndexError:- it is raised while accessing an element in a sequence like list using index that is outside the valid index range.

```python
my_list = [10, 20, 30, 40, 50]

try:
    element = my_list[6]  
except IndexError as e:
    print(f"Error: {e}")
```

## Q5. Explain ImportError. What is ModuleNotFoundError?

--Import error is an exception class that occurs when there is an issue with the importing a module. it signifies that there is a problem while trying to import a module or component.

for Example:-
1. ModuleNotFoundError:- it is raise when python interpretor cannot locate the module which is being imported. this exception occurs when module does not exist or not installed on system.

```python
try:
    import non_existent_module  
except ModuleNotFoundError as e:
    print(f"Error: {e}")
```

##  Q6. List down some best practices for exception handling in python.

1. Catch Specific exception types rather than using a generic except clause whenever possible.

2. Handle Exception by providing informative error messages and taking appropriate actions to recover from the error.

3. Wrap potential error-prone code in try and except block this isolates that code might raise exceptions.

4. use logging to log exceptions and error messages.
