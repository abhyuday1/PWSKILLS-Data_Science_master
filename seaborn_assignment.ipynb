## 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

1. Scatter Plot:Scatter plots are used to visualize the relationship between two continuous variables. They can help identify patterns, trends, and outliers in the data.

2. relplot:Visualizing relationships between two continuous variables using scatter plots.

3. pairplot:Visualizing pairwise relationships between multiple continuous variables.
4. jointplot: It's used to explore the relationship between two variables along with their individual distributions.
5. distplot:-distplot is used to visualize the univariate distribution of a single continuous variable

## 2  Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x = "timepoint" and y = "signal" for different events and regions.

```python
import seaborn as sns
from matplotlib import pyplot as plt
#import matplotlib.pyplot as plt
data= sns.load_dataset('fmri')
sns.relplot(x="timepoint",y='signal',hue='event',style='region',kind='line',data=data )
plt.show()
```

## Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x ='pclass', y = 'age' and y = 'fare'.

```python


import seaborn as sns
import matplotlib.pyplot as plt

titanic_data=sns.load_dataset("titanic")
plt.subplot(1,2,1)
sns.boxplot(x='pclass',y='age',data=titanic_data)
plt.subplot(1,2,2)
sns.boxplot(x='pclass',y='fare',data=titanic_data)
plt.tight_layout()
plt.show()
```

## Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue parameter for the 'cut' column of the diamonds dataset.

```python
import seaborn as sns
import matplotlib.pyplot as plt

diamond_data=sns.load_dataset("diamonds")
sns.histplot(data=diamond_data,x='price',hue='cut')
plt.show()
```

## Que 5: Use the "iris" dataset from seaborn to plot a pair plot. Use the hue parameter for the "species" column of the iris dataset.

```python
import seaborn as sns
import matplotlib.pyplot as plt

iris=sns.load_dataset('iris')
sns.pairplot(iris,hue='species')
plt.show()
```
