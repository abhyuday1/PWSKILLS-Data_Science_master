## Q1. What is Flask Framework? What are the advantages of Flask Framework?

--Flask is a micro web framework for Python.Flask provides the essentials for web development without imposing strict patterns or dependencies

#### Advantages of Flask Framework:
1. Flexibility:- it provides the core components for web development but allows developers to choose and integrate additional libraries and tools as needed.
2. Lightweight: Flask has a small codebase, which results in faster startup times and reduced memory usage compared to larger web frameworks. This makes it well-suited for small to medium-sized applications.
3. Rapid Development: Due to its simplicity and minimal setup, Flask is suitable for rapid prototyping and development. It allows developers to focus on building features rather than dealing with complex configurations.
4. RESTful API Development: Flask is commonly used for building RESTful APIs due to its simplicity and support for handling HTTP requests and responses.

## Q3. What is App routing in Flask? Why do we use app routes?
-- App routing refers to the process of defining URL routes and specifying the corresponding view functions that should handle incoming HTTP requests to those routes. Routes define the paths that users can access in a web application, and they play a crucial role in how Flask processes and responds to different requests. Flask uses the @app.route decorator to define routes

#### Why Do We Use App Routes in Flask?

1. App routes help organize web application by defining clear URL patterns and associating them with specific functionality.
2. Routes are essential for directing incoming requests to the appropriate view functions that generate responses.
3. Flask's route system allows for URL generation based on route names, which promotes clean and consistent URL construction.

## Q4. Create a “/welcome” route to display the welcome message “Welcome to ABC Corporation” and a “/” route to show the following details:
Company Name: ABC Corporation
Location: India
Contact Detail: 999-999-9999


```python
from flask import Flask
app1=Flask(__name__)

@app1.route('/welcome')
def welcome():
    return "Welcome to ABC Corporation"

@app1.route('/')
def company_details():
    return '''
    Company Name: ABC Corporation
    Location: India
    Contact Detail: 999-999-9999
    '''
    
if __name__=="__main__":
    app1.run()
    
```
