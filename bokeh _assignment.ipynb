## Q1. How can you create a Bokeh plot using Python code?

```python
##first we have to install the Bokeh library
!pip install bokeh
```

```python
# ini second step we have to import the necessary modules
from bokeh.plotting import figure, show,output_file
from bokeh.sampledata.iris import flowers
import bokeh.io
bokeh.io.output_notebook()
```

```python
flowers
```

```python
p=figure(title='test flower',x_axis_label='petal_length',y_axis_label="petal_width")

p.circle(flowers['petal_length'],flowers['petal_width'])
show(p)
```

## Q2. What are glyphs in Bokeh, and how can you add them to a Bokeh plot? Explain with an example.

--Glyphs are used to visualize the data in various ways, such as circles, squares, lines, bars

```python

x = [1, 2, 3, 4, 5]
y = [6, 7, 2, 4, 5]

p = figure(title="Line Plot", x_axis_label="X-axis", y_axis_label="Y-axis")


p.line(x, y, line_width=2, color="green")


show(p)
```

## Q3. How can you customize the appearance of a Bokeh plot, including the axes, title, and legend?

--we can customize the appearance of a Bokeh plot, including the axes, title, and legend, by modifying various attributes of the figure object

```python
p=figure(title='test flower',x_axis_label='petal_length',y_axis_label="petal_width")

p.circle(flowers['petal_length'],flowers['petal_width'],legend_label='Petal points')
show(p)
```

## Q4. What is a Bokeh server, and how can you use it to create interactive plots that can be updated in real time?

--A Bokeh server is a Python process that allows  to create and deploy interactive, real-time data applications and dashboards with Bokeh. Bokeh provides a server framework that enables to build dynamic web applications with interactive plots, widgets, and data updates without needing to write a separate backend server. With a Bokeh server, we can create applications that respond to user input and update the visualizations in real-time, making it a powerful tool for building data-driven web applications.
